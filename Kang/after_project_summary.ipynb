### 2.1 다중공선성(VIF) 분석

```python
#선택한 데이터 이외 데이터 제거
feature = ["log1p_dist", "pickup_longitude", "dropoff_longitude", "month", "hour", "day", "vendor_id", "log1p_y"]
not_feature = [column for column in train.columns if column not in feature]
feature_a = train.drop(columns = not_feature)
```

```python
vif_test = feature_a.iloc[ : , : -1]

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(vif_test.values, i) for i in range(vif_test.shape[1])]
vif["features"] = vif_test.columns
vif
```
